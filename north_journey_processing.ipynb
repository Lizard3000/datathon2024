# Extract Table Data From Journey North Website

### 1. change the url

### 2. change the csv file name that will be saved

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://journeynorth.org/sightings/querylist.html?season=&map=monarch-adult-fall&year=1997&submit=View+Data"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

text = soup.find("table")

table = soup.select("table")

table = table[0]

table_df_list = pd.read_html(str(table))

table_df = table_df_list[0]
table_df
```

```python
table_df.to_csv("Monarch_Adult_Sighted_97.csv")
```
